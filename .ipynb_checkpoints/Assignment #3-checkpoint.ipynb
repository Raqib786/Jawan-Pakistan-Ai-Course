```python
# make a calculator using python with addition ,subtraction,multiplication,division and power

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
print("****************************************")
print("\t Select operand")
print("****************************************")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division ")
print("5. Power ")
print("****************************************")
choice = input("Enter your choice : ")


if choice == '1':
    print(num1 ,"+", num2 ,"Addition is :",num1 + num2)
elif choice == '2':
    print(num1 ,"-", num2 ,"subtraction is :",num1 - num2)
elif choice == '3':
    print(num1 ,"*", num2 ,"multiplication is :",num1 * num2)
elif choice == '4':
    if num2 != 0:
        print(num1 ,"/", num2 ,"division is :",num1 / num2)
    else:
        print("Error! Division by zero is not allowed.")
elif choice == '5':
    print(num1 ,"^", num2 ,"Power is :", num1 ** num2)
else:
    print("Invalid Operand")
```

```python
# Write a program to check if there is any numeric value in list using for loop.

list = [1,2,"asad",4,5,6,7,8]

for check in list:
    if isinstance(check, (int, float)):  # check if item is a number
        
        print("Number Exist in this list")
        break  # stop loop once a numeric value is found
else:
    print("Number not Exist in this list")
    
    
```

```python
# Write a python script to add a key to a dictionary.

my_dict = {"name": "Ali", "age": 25}

# Adding a new key-value pair
my_dict["city"] = "Karachi"

print(my_dict)
```

```python
# Write a program to sum all the numeric item in a dictionary.

my_dict = {"num": 24, "name": "Jack", "num2": 80,"num3":60}

total = 0

for value in my_dict.values():
    if isinstance(value, (int, float)):  # check if value is numeric
        total += value

print("Sum of numeric items in this list :", total)
```

```python
# Write a program to identify duplicate values from the list.
list = [1,2,"asad",4,5,6,7,8,3,5,3,4,4]
unique = []
duplicate = []
for digits in list:
    if digits not in unique:
        unique.append(digits)
    elif digits not in duplicate:
        duplicate.append(digits)
print("Duplicate values:", duplicate)

        
```

```python
# Write a program to check if a given key already exists in a dictionary.

dict_name = {"name": "Ali", "age": 25, "city": "Karachi","street":9}

# Key to check
check_key = "city"

# Check if key exists
if check_key in dict_name:
    print(check_key,"is already exists in the dictionary.")
else:
    print(check_key,"is does not exist in the dictionary.")
```
